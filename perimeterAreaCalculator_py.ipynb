```python
#Area and perimeter calculator:
print('Please input the width')
width = int(input())
print('Please input the length')
length = int(input())

def area(length,width):
  area = length * width
  print(f'The total area is {area}')
  return area

def perimeter(length,width):
  perimeter = (length * 2) + (width * 2)
  print(f'The total perimeter is {perimeter}')
  return perimeter

print(area(width,length))
print(perimeter(width,length))
```
